torch.dot函数对向量x求内积，使用grad.zero_()对之前求得的梯度清零，否则会累积

```python
x.grad.zero_()
y = 2 * torch.dot(x, x)
y
```

反向传播可以计算出y对$\vec x$每一个分量的梯度，并把结果储存在$\vec x$的grad成员里

```python
y.backward()
x.grad, x.grad == 4 * x
```

### 生成带噪声的数据集——synthetic_data

根据给出的权重和偏置，加上噪声，得到数据集

```python
# 权重, 偏置, 样本数
def synthetic_data(w, b, num_examples):
    # w是个len(w)长的向量, b是个标量
    x = torch.normal(0, 1, size = (num_examples, len(w))) # x: 均值为0, 方差为1的随机数, n个样本, w列
    y = torch.matmul(x, w) + b # x的每一行和w求内积, 再加上偏置
    y += torch.normal(0, 0.01, y.shape) # 加随机噪音, y此时是一个行向量
    print(x.shape, w.shape, y.shape)
    return x, y.reshape((-1, 1)) # y.shape为[1000, 1]

true_w = torch.tensor([2, -3.4])
true_b = 4.2
features, labels = synthetic_data(true_w, true_b, 1000)

print('features:', features[0], '\nlabel:', labels[0], '\nw.shape:', true_w.shape)
d2l.set_figsize()
d2l.plt.scatter(features[:, 1].detach().numpy(), labels.detach().numpy(), 1)
```

### 读取小批量——data_iter

将生成的features(x)，labels(y)，按照给定的batch_size划分，每次输出一个批量的数据

使用yield的好处之一，是可以节省内存，当需要迭代的时候才生成一次

```python
def data_iter(batch_size, features, labels):
    # 生成一个样本序号的list, 并将其打乱, 方便后续访问
    num_example = len(features)
    indices = list(range(num_example))
    random.shuffle(indices)
    
    # 从0开始每次取batch_size个样本
    for i in range(0, num_example, batch_size):
        # 取batch_size个样本, 最后不够取最小即可, 用切片的下标索引方式取样本
        batch_indices = indices[i: min(i+batch_size, num_example)]
        
        # batch_indices是一组随机下标序列, yield返回一个可迭代对象
        yield features[batch_indices], labels[batch_indices] 
        
batch_size = 10
for x, y in data_iter(batch_size, features, labels):
    print(x, '\n', y)
    # break表示终止迭代, 否则会一直进行到迭代器结束为止
    break
```

### 定义初始化参数

```python
w = torch.normal(0, 0.01, size= (2,), requires_grad=True)
b = torch.zeros((1,), requires_grad = True)
w, w.shape
```

### 定义模型——linreg

```python
def linreg(x,w,b):
    return torch.matmul(x,w)+b
```

### 定义损失函数——squared_loss

```python
'''
net = linreg # 网络结构
loss = squared_loss # 损失函数
l = loss(net(x,w,b),y)
由linreg(x,w,b)=torch.matmul(x,w)+b给出预测值y_hat
此时的y_hat：n行两列的x乘以长为2的向量w，此时[[...],[...],[...],...]的里层括号被消解，转换为一个标量
y_hat的shape为：长为n的向量
y的shape为：size = (n,1)，需要将y的形状reshape为y_hat的形状
返回的结果为一个长度为n的向量，每一位为预测的y和实际的y之间的误差
'''
def squared_loss(y_hat, y):
    # 均方误差
    # print(y_hat.shape, y.shape)
    return (y_hat - y.reshape(y_hat.shape))**2 / 2
```

y_hat和y的size不一样，用到了广播机制

```python
a = np.array([1,2,3]) # a.shape = 3 -> 1 * 3
b = np.array([[1],[2],[3]]) # b.shape = 3 * 1
a_ = np.array([[1,2,3],[1,2,3],[1,2,3]]) # 将a的第一维进行广播, 复制到3行 
b_ = np.array([[1,1,1],[2,2,2],[3,3,3]]) # 将b的第二维进行广播, 复制到3列
a - b, a_ - b_
```

### 定义优化算法——sgd

```python
'''
params为需要求梯度参数，此时传进的为[w,b]
'''
def sgd(params, lr, batch_size):
    # 小批量随机梯度下降
    # 更新不计算梯度, no_grad
    with torch.no_grad():
        # 对每个参数进行梯度更新
        for param in params:
            param -= lr * param.grad / batch_size
            param.grad.zero_()
```

### 训练过程 

```python
lr = 0.03 # 学习率
num_epochs = 3 # 总循环次数
net = linreg # 网络结构
loss = squared_loss # 损失函数

for epoch in range(num_epochs):
    # 一轮对每个小批量的batch_size都进行更新, 循环结束的条件即是迭代器结束
    for x, y in data_iter(batch_size, features, labels):
        
        # linreg(x,w,b)生成y_hat
        l = loss(net(x,w,b), y)
        
        # l做反向传播, 这样(w,b)能获得梯度
        # 此时对l求和, l的每一位都有个对[w,b]的运算: y_hat = wx+b, 求和后即将w的梯度求和, 将b的梯度求和
        # l = (wx+b-y)^2, w和b的梯度和数据里的x和y有关
        l.sum().backward()
        
        sgd([w,b], lr, batch_size) # 参数梯度更新
        
    with torch.no_grad():
        train_l = loss(net(features, w, b), labels)
        print(f'epoch {epoch + 1}, loss {float(train_l.mean()):f}')
```

python的格式化输出，`f''`表示`` 里的内容用格式化输出，{a}会自动的引用变量a的值，{a:10.4f}表示将a用浮点数的形式输出出来，其占10个宽度，保留小数点后4位

```python
a = 3.1415926221231
print(f'{a},{a:10.4f}')
```

## pytorch实现

### 产生数据集 

```python
true_w = torch.tensor([2, -3.4])
true_b = 4.2
features, labels = d2l.synthetic_data(true_w, true_b, 1000)
```

### 划分batch数据 

```python
def load_array(data_arrays, batch_size, is_train = True):
    # 构建pytorch数据迭代器
    dataset = data.TensorDataset(*data_arrays)
    return data.DataLoader(dataset, batch_size, shuffle = is_train)

batch_size = 10
data_iter = load_array((features, labels), batch_size)

next(iter(data_iter)), type(data_iter), list(data_iter)
```

### 定义模型 

```python
net = torch.nn.Sequential(torch.nn.Linear(2, 1))
```

### 初始化参数 

```python
net[0].weight.data.normal_(0, 0.01) # 正太分布初始化
net[0].bias.data.fill_(0)
net[0].weight, net[0].bias
```

### 定义误差和优化方法 

```python
# 均方误差
loss = torch.nn.MSELoss()
# 传入net的参数，设置学习率
trainer = torch.optim.SGD(net.parameters(), lr = 0.03)
```

### 训练模型 

```python
num_epochs = 3
for epoch in range(num_epochs):
    for x, y in data_iter:
        l = loss(net(x), y)
        trainer.zero_grad()
        l.backward()
        trainer.step()
    l = loss(net(features), labels)
    print(f'epoch{epoch + 1}, loss{l:f}')
```

```python
def load_data_fashion_mnist(batch_size, resize = None):
    trans = [transforms.ToTensor()]
    
    # 为trans赋予能修改图片大小的功能
    if resize:
        trans.insert(0,transforms.Resize(resize))
        
    # trans是个列表，通过Compose将列表中的多个操作合并
    trans = transforms.Compose(trans)
    
    mnist_train = torchvision.datasets.FashionMNIST(root = './dataset/FashionMnist', train = True, transform = trans, download = True)
    mnist_test = torchvision.datasets.FashionMNIST(root = './dataset/FashionMnist', train = False, transform = trans, download = True)
    
    return (data.DataLoader(mnist_train, batch_size, shuffle = True, num_workers = get_dataloader_workers),
            data.DataLoader(mnist_test, batch_size, shuffle = False, num_workers = get_dataloader_workers))
```

### 实现softmax 

```python
def softmax(x):
    x_exp = torch.exp(x)
    # 二维矩阵(n,m)：对列求和得到(1,m)，axis=0；对行求和得到(n,1)，axis=1
    # x_exp是一个batch_size * 784的矩阵，对每一行求和，axis=1
    sum_raw = x_exp.sum(1, keepdim = True)
    # 使用广播机制，对每一行的各个x(即最终输出的10个值)求出"概率"
    return x_exp / sum_raw

x = torch.normal(0, 1, (2, 5))
x_prob = softmax(x)
x, x_prob, x_prob.sum(1, keepdim=True)
```

### 设置模型 

```python
def net(x):
    # x.shape:(batch_size, 784)
    # w.shape:(784, 10)
    # matmul:(batch_size, 10)
    # b:(10)
    # 通过广播机制将每个batch_size都和b相加
    return softmax(torch.matmul(x.reshape(-1, w.shape[0]), w) + b)

a = torch.arange(10).reshape(5,2)
b = torch.tensor([1., 2.])
a, b, a + b
```

### 实现交叉熵 

注意到，pytorch的tensor可以通过列表来索引其中的部分元素

```python
y = torch.tensor([0, 2])
y_hat = torch.tensor([[0.1, 0.3, 0.6], [0.3, 0.2, 0.5]])

# 通过列表索引，筛选出y_hat里和label值对应的概率，可以用于后续求交叉熵
y_hat[[0, 1], y] # 等价于y[0][0], y[1][2]即第0组选第0个元素，第1组选第2个元素
```

```python
def cross_entropy(y_hat, y):
    return -torch.log(y_hat[range(len(y_hat)), y])

cross_entropy(y_hat, y)
```

### 与真实y进行比较，求正确率 

```python
def accuracy(y_hat, y):
    # argmax返回列表里最大元素的下标，可以按指定维度进行比较
    y_predict = y_hat.argmax(axis = 1)
    cmp = y_predict.type(y.dtype) == y
    return float(cmp.sum())
    
accuracy(y_hat, y) / len(y)
```

### 计算模型的精度 

创建累加器

```python
class Accumulator:
    def __init__(self, n):
        self.data = [0.0 for i in range(n)]
        self.length = n
        
    def add(self, *args):
        self.data = [self.data[i] + args[i] for i in range(self.length)]
        
    def reset(self):
        self.data = [0.0 for i in range(self.length)]
        
    # 魔法方法，可以通过[]访问叠加器里的某个idx元素    
    def __getitem__(self, idx):
        return self.data[idx]
```

```python
def evaluate_accuracy(net, data_iter):
    # 将模型设置为评估模式不计算梯度
    if isinstance(net, torch.nn.Module):
        net.eval() 
    metric = Accumulator(2)
    for x,y in data_iter:
        metric.add(accuracy(net(x), y), y.numel())
    return metric[0] / metric[1]

evaluate_accuracy(net, test_iter)
```
