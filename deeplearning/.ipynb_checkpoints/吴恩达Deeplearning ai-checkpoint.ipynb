# 简介

## 什么是神经网络

ReLU: rectified linear unit线性修正单元，例如房价预测线性模型，不可能低于0，故在小于某个数时，取常数值；简单的神经网络里的每个节点，可以看作输入经过某一种ReLU变换

<img src='.\Image\吴恩达Deeplearning\简单神经网络.png' width=500>

其中的每个节点可能有其实际含义，但神经网络并不关心，它会自动计算输入层和输出层中的隐藏层的参数

## 用神经网络进行监督学习

常用神经网络：
+ 标准神经网络：Standard NN
+ 卷积神经网络CNN：通常用于图像处理，Convolution Neural Network
+ 循环神经网络RNN：通常用于序列数据，例如语音——随着时间序列变化，语音——一个个单词序列，Recurrent Neural Network

<img src='.\Image\吴恩达Deeplearning\神经网络种类.png' width=500>

数据的表示：
+ 结构化数据：表、数据库
 + 房价预测——地理位置，周边设施……
 + 广告推荐——用户年龄，浏览偏好……
+ 非结构化数据：图像、音频

## 为什么深度学习会兴起

相比于其它方法的优势：
+ 依赖数据规模——大数据的发展
+ 依赖训练规模——CPU/GPU算力的提高

算法的改进：从sigmoid（梯度较小）转变为relu函数，使得神经网络不会随着层数增大而使得前面的层训练变化较小

------
+ 对于ReLu函数的叠加，它看上去更像线性函数的分段

```python
def ReLu(x):
    return np.maximum(0, x)

import matplotlib.pylab as plt
import numpy as np

x = np.arange(-20,20,1)
y = ReLu(ReLu(-ReLu(ReLu(ReLu(ReLu(x)-2)+5)-13)+41)-17)
plt.plot(x, y)
plt.show()
```

+ 而对于ReLu函数的求和，则能模拟出曲线函数的形状

```python
def ReLu(x):
    return np.maximum(0, x)

import matplotlib.pylab as plt
import numpy as np

x = np.arange(-20,20,1)
y = ReLu(-2*x+1)+ReLu(3*x-3)+ReLu(-2*x-5)+ReLu(4*x-100)
plt.plot(x, y)
plt.show()
```
